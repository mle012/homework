```python
import pandas as pd
```

```python
#STEP 1 Create a data frame

data = {
    'name': ['Nate', 'John', 'Jane', 'Mandy'],
    'age': [39, 23, 31, 25],
    'sex': ['m', 'm', 'f', 'f'],
    'email': ['nate@gmail.com', 'john@gmail.com', 'jane@gmail.com', 'mandy@gmail.com'],
    'purchases': [34,12,42,8]
}   
df = pd.DataFrame(data)
print (df)
```

```python
#STEP 2: RENAME THE 'SEX' COLUMMN TO GENDER
df = df.rename(columns={'sex': 'gender'})
print (df)
```

```python
#Step 3 add a new column named active and set all the values to true
df['active'] = True
print (df)
```

```python
#step 4 print only columns 'name' and 'email' 
print (df[['name', 'email']])
```

```python
#step 5, add 5 more purchases to every entry in the purchases column and print the data frame
df['purchases'] = df['purchases'] + 5
print(df)
```

```python
# step 6 print the filtered dataframe to include onlu rows where purchases is greater than 20
filtered_df = df[df['purchases'] > 20]
print(filtered_df)
```

```python
# step 7, change all m to male, f to female in gender column and print the data frame
df.loc[df["gender"] == "m", "gender"] = "male"
df.loc[df["gender"] == "f", "gender"] = "female"
print(df)
```
