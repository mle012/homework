```python
# step one: 5 variables, types, and print each one
name = "Maria"          # string
age = 21         # integer
grade = 99.9     # float
is_student = True  # boolean
favorite_color = "Green" # string

print(name)
print(age)
print(grade)
print(is_student)
print(favorite_color) 
```

```python
# step two: reassign each variable to the same value, use type hinting to declare data type for each variable
name: str = "Maria"
age: int = 21
grade: float = 99.9
is_student: bool = True
favorite_color: str = "Green"
```

```python
# step three: reassign the name variable, uppercase it, and print it
name = name.upper()
print(name)
```

```python
#step four: add age variable to the grade variable. print the result 
result = age + grade
print(type(result))
```

```python
#step five: create dictionary named person, include variable in step two, print it
person = {
    "name": name,
    "age": age,
    "grade": grade,
    "is_student": is_student,
    "favorite_color": favorite_color
}
print(person)
```

```python
#step six: use f string, print sentence including all variables from person dictionary
print(f"My name is {person['name']}, I am {person['age']} years old. My favorite color is {person['favorite_color']}. I will get a {person['grade']} in this class.")
```
